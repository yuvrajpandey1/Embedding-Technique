### Ollama

* Ollama supports embedding models, making it possible to build retrieval argumented generation (RAG) applications that combine text prompts with existing documents or other data.

```python
pip install langchain
```

```python
pip install langchain-community
```

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma:2b")    ## By default it use llama2 model
)
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
        "Itni Shiddat Se Maine Tumhe Paane Ki Koshish Ki Hai,Ki Har Zarre Ne Mujhey Tumse Milane Ki Saazish Ki Hai",
        "22 Tak Padhai, 25 Pe Naukri, 26 Pe Chokri, 30 Pe Bachhe, 60 Pe Retirement.... aur Phir Maut Ka Intezaar... Dhat Aaisi Ghisi Piti Life Thodi Jeena Chahta Hoon."
    ]
)
```

```python
r1[0]
```

```python
len(r1[0])
```

```python
embeddings.embed_query("Tumse Milane Ki Saazish Kisney Ki Hai?")
```
